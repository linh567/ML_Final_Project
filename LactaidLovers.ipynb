# 5. Data Cleaning Plan

Based on what we discovered, our cleaning steps will include:

- Handling redundant features

- Drop Feed_Quantity_lb (kg column is sufficient)

- Handling date columns

- Drop if irrelevant OR extract useful components (season, month)

- Combining vaccination features

- Create a single Total_Vaccinations or binary Any_Vaccinated

- Categorical cleanup

- Convert Farm_ID to string/category

- Handling missing values

- Numeric: impute with median

- Categorical: impute with mode

- Domain-specific decisions where necessary

- Outlier handling

- Review milk yield and feed quantity for extreme values

```python
# Make a working copy
clean_df = df.copy()

# 1. Drop redundant columns
if "Feed_Quantity_lb" in clean_df.columns:
    clean_df = clean_df.drop(columns=["Feed_Quantity_lb"])

# 2. Convert categorical columns
clean_df["Farm_ID"] = clean_df["Farm_ID"].astype(str)

# 3. Combine vaccination columns (example – adjust to actual column names)
vaccine_cols = [col for col in clean_df.columns if "Vaccine" in col]
if vaccine_cols:
    clean_df["Total_Vaccines"] = clean_df[vaccine_cols].sum(axis=1)

# 4. Drop date column if irrelevant
if "Date" in clean_df.columns:
    clean_df = clean_df.drop(columns=["Date"])

clean_df.head()
```

# 6. Missing Value Imputation

We impute:

Numeric → median

Categorical → mode

```python
from sklearn.impute import SimpleImputer

num_cols = clean_df.select_dtypes(include=np.number).columns
cat_cols = clean_df.select_dtypes(exclude=np.number).columns

num_imputer = SimpleImputer(strategy="median")
cat_imputer = SimpleImputer(strategy="most_frequent")

clean_df[num_cols] = num_imputer.fit_transform(clean_df[num_cols])
clean_df[cat_cols] = cat_imputer.fit_transform(clean_df[cat_cols])

clean_df.isna().sum()
```

# 8. Negative values in dataset
We examine the number of negative values in our dataset. There are three features with negative values: rumination time (hrs), ambient temperature (celsius), and milk yield (L). Below we walk through our reasoning for cleaning and processing of these features.

1. Rumination time (hrs): since hours cannot be negative, we know that these values are invalid. More than half of the data contain invalid values, so we decide to drop this column
2. Ambient temperature (celsius): temperature can be negative. A quick google search tells us that cows can withstand temperatures as low as -28 celsius. The lowest temperature in the dataset is -11.7 celsius, so we can keep these values
3. Milk yield (L): milk yield cannot be negative. Since there are only 74 records with invalid values, we decide to drop these rercords

```python
from sklearn.metrics import mean_squared_error, r2_score
negative_counts = (clean_df[num_cols] < 0).sum()
print(negative_counts)
# data cleaning
dropped_df = clean_df.drop(columns=["Rumination_Time_hrs", "HS_Vaccine", "BQ_Vaccine"])
dropped_df = dropped_df[dropped_df["Milk_Yield_L"] >= 0]
```

# 8. Further Data Prep

To prep our data for analysis, we will need to perform one hot encoding, ordinal mapping, aswell as drop Cattle_Id

```python
# Drop the ID columns since they are identifiers
dropped_df = dropped_df.drop(columns=["Cattle_ID", "Farm_ID"])

#Through visual inpsection, we see that these columns should be one hot encoded
one_hot_cols = ['Breed', 'Climate_Zone', 'Management_System', 'Feed_Type']
dropped_df = pd.get_dummies(dropped_df, columns=one_hot_cols, drop_first=True)

#Lactation stage only has 3 possible values and they are in order, so we use ordinal mapping
ordinal_col = 'Lactation_Stage'
ordinal_mapping = {'Early': 0, 'Mid': 1, 'Late': 2}
dropped_df[ordinal_col] = dropped_df[ordinal_col].map(ordinal_mapping)
```

# 9. Initial Implementation--RandomForestRegressor

Since milk yield is continous, our initial approach is to use ensemble with random forest regressor. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Stratified sampling
sample_size = 1000
# Create bins for milk yield to ensure representative sampling
dropped_df['yield_bin'] = pd.qcut(dropped_df['Milk_Yield_L'], q=5, labels=False, duplicates='drop')

sampled_df = dropped_df.groupby('yield_bin', group_keys=False).apply(
    lambda x: x.sample(min(len(x), sample_size // 5), random_state=42),
    include_groups=False
)
# Drop the temporary binning column
#sampled_df = sampled_df.drop(columns=['yield_bin'])

y = sampled_df['Milk_Yield_L']
X = sampled_df.drop(columns=['Milk_Yield_L'])
# PCA
scaler = StandardScaler()
print(X.isna().sum())
X_scaled = scaler.fit_transform(X)
# Fit PCA to see explained variance
pca_full = PCA()
pca_full.fit(X_scaled)
n_components_95 = np.argmax(np.cumsum(pca_full.explained_variance_ratio_) >= 0.95) + 1
# Apply PCA with optimal components (95% variance)
pca = PCA(n_components=n_components_95)
X_pca = pca.fit_transform(X_scaled)
print("pca finished")
rf = RandomForestRegressor(random_state=42)
param_grid = {'n_estimators': [50, 100, 150]}
gs = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    verbose=3,   # <--- prints after each fit
    n_jobs=-1
)
gs.fit(X_pca, y) # <--- Fit GridSearchCV first

# 2. Get the best estimator from the search
best_model = gs.best_estimator_

# 3. Use cross_val_predict with the best, *already-tuned* estimator
# This will perform cross-validation on the best_model to get OOF predictions.
preds = cross_val_predict(best_model, X_pca, y, cv=5)

# Evaluate
mse = mean_squared_error(y, preds)
rmse = np.sqrt(mse)
r2 = r2_score(y, preds)
print(f"RMSE: {rmse:.3f}")
print(f"R²: {r2:.3f}")
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Stratified sampling
sample_size = 200
dropped_df['yield_bin'] = pd.qcut(
    dropped_df['Milk_Yield_L'], q=5,
    labels=False, duplicates='drop'
)

sampled_df = dropped_df.groupby('yield_bin', group_keys=False).apply(
    lambda x: x.sample(min(len(x), sample_size // 5), random_state=42),
    include_groups=False
)

# Prepare X and y
y = sampled_df['Milk_Yield_L']
X = sampled_df.drop(columns=['Milk_Yield_L'])

# PCA
scaler = StandardScaler()
print(X.isna().sum())
X_scaled = scaler.fit_transform(X)
pca_full = PCA()
pca_full.fit(X_scaled)
n_components_95 = np.argmax(
    np.cumsum(pca_full.explained_variance_ratio_) >= 0.95
) + 1

pca = PCA(n_components=n_components_95)
X_pca = pca.fit_transform(X_scaled)

print("pca finished")

# -----------------------
# KNN MODEL SECTION
# -----------------------

# Define the model
knn = KNeighborsRegressor()
param_grid = {
    'n_neighbors': [2, 3, 4, 5, 7, 9, 11, 15, 20]
}
gs = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5
)
gs.fit(X_pca, y)

# Best model
best_model = gs.best_estimator_
print("Best k:", gs.best_params_['n_neighbors'])

# Cross-validated predictions
preds = cross_val_predict(best_model, X_pca, y, cv=5)

# Metrics
mse = mean_squared_error(y, preds)
rmse = np.sqrt(mse)
r2 = r2_score(y, preds)

print(f"RMSE: {rmse:.3f}")
print(f"R²: {r2:.3f}")
```
